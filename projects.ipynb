```python
import pandas as pd
import glob

# Path to all CSV files
csv_files = glob.glob("Action.csv")

# List to store DataFrames
dfs = []

for file in csv_files:
    # Extract genre name from file name (assuming 'comedy.csv' -> 'comedy')
    genre = file.split("/")[-1].split(".")[0].capitalize()
    
    # Read CSV
    df = pd.read_csv(file)
    
    # Add 'Genre' column
    df['Genre'] = genre
    
    # Append to list
    dfs.append(df)

# Merge all DataFrames
merged_df = pd.concat(dfs, ignore_index=True)

# Save merged file
merged_df.to_csv("actions.csv", index=False)

print("Merged CSV with Genre column saved successfully!")
```

```python
import pandas as pd
import re

# Load your CSV
df = pd.read_csv("C:\\Users\\ELCOT\\Documents\\IMDB_movies.csv")

# Function to convert duration
def convert_to_minutes(duration_str):
    if pd.isna(duration_str):
        return None
    hours = re.search(r'(\d+)h', duration_str)
    minutes = re.search(r'(\d+)min', duration_str)

    total_minutes = 0
    if hours:
        total_minutes += int(hours.group(1)) * 60
    if minutes:
        total_minutes += int(minutes.group(1))
    return total_minutes

# Apply conversion
df['Duration'] = df['Duration'].apply(convert_to_minutes)

# Save back to CSV
df.to_csv("IMDB.csv", index=False)
```

```python
import sqlite3
import pandas as pd

# Define the database file
db_file = "movie.db"

# Connect to SQLite3
conn = sqlite3.connect(db_file)
cursor = conn.cursor()

# Create a table
cursor.execute("""
CREATE TABLE IF NOT EXISTS movies (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    genre TEXT,
    rating REAL,
    votes INTEGER,
    duration INTEGER
)
""")

conn.commit()
conn.close()
```

```python
def save_to_sqlite(df, db_file="movie.db"):
    # Connect to SQLite
    conn = sqlite3.connect(db_file)

    # Store DataFrame into SQLite
    df.to_sql("movie", conn, if_exists="replace", index=False)

    # Close connection
    conn.close()
    print("Data successfully stored in SQLite!")

# Example usage
df = pd.read_csv("IMDB_movies.csv")  # Load merged dataset
save_to_sqlite(df)
```
